## Problem
V projektni nalogi bomo na podlagi prometnih nesreč na slovenskih cestah med letoma 2009 in 2022 naredili pripravo, obdelavo in analizo izbranih podatkov, ki smo jih na tej temi pridobili iz zbirke podatkov Ministrstva za javno upravo na spletni strani OPSI. 
Za ostale potrebe kot je normalizacija in podobno smo uporabili druge podatkovne zbirke, tudi iz OPSI in še iz zbirke podatkov Statističnega urada Republike Slovenije na spletni strani SiStat.



## Podatki
Datoteke s podatki združujejo prometne nesreče in osebe, ki nastopajo v prometnih nesrečah kot povzročitelji ali udeleženci. V datotekah so zajete vse prometne nesreče, pri ogledu katerih je sodelovala policija.

Podatke smo jemali iz:
* 14 zbirk s podatki o nesrečah, ena za vsako leto od leta 2009 do 2022,
* Zbirka podatkov o prebivalstvu občin glede na starost od leta 2009 do 2022,
* 13 zbirk podatkov o prometnih obremenitvah državnega cestnega omrežja od leta 2009 do 2021.

Glavna zbirka s podatki o nesrečah, vsaka vsebuje po 35 atributov, v katerih vsaka vrstica predstavlja eno osebo, ki je bila udeležena neke prometne nesreče. Vsak atribut ima svojo vrednost ali opis, ki smo ga uporabili za analizo in vizualizacijo podatkov.
Zbirko o prebivalstvu občin glede na starost, podatke od gostoti javnega prometa po regijah in podatke o prometnih obremenitvah državnega cestnega omrežja pa smo uporabili za normalizacijo rezultatov in prikazovanje drugih informativnih  grafov.

Glavna zbirka podatkov je zapisana v CSV formatu, ločeni s podpičji.

Problemi, s katerimi smo se soočali pri obdelavi podatkov ter naše rešitve:
* Šumnik »Č« je bil nepravilno kodiran
  * Vse nepravilne pojave smo nadomestili s programom.
* Nekatere vrednosti so prazne 
  * Te ignoriramo ali jim dodelimo privzete vrednosti.
* Težje pridobivanje nekaterih podatkov,
  * Zapis v drugačnih formatih, ročni izpis iz nekaterih spletnih strani.
* Dokument ni bil formatiran v utf-8 kar je povzročalo probleme pri branju
  * Re-formatiranje dokumentov v pravi format
* Pretvarjanje koordinatnih sistemov v WGS84 (se uporablja tudi pri Googlu):
  * 2009-2012: Gauß-Krüger koordinatni sistem
  * 2013-2022: D96 koordinatni sistem

Za obdelavo in prikazovanje podatkov smo uporabili naslednje knjižnice: 
* Sklearn
* Pandas
* Numpy
* SciPy
* StatsModels
* Matplotlib
* Shapefile
* Imageio 





## Rezultati in ugotovitve

Začeli smo z vprašanjem o deležu nesreč povzročenih zaradi neprilagojene hitrosti in nepravilne smeri vožnje, kjer so bili vozniki alkoholizirani.



Iz grafa lahko razberemo, da čeprav alkoholizirani vozniki povzročijo 20 – 25 % nesreč zaradi neprilagojene hitrosti in nepravilne smeri vožnje, moramo upoštevati, da alkoholizirani vozniki povzročijo približno 12.4% vseh nesreč. To pomeni, da so alkoholizirani vozniki bolj verjetni, da povzročijo enega od teh dveh tipov nesreč kot pa ne alkoholiziran vozniki.

## Napovedovanje
Za napovedovanje smo uporabljali klasifikacijo, ki napoveduje vzrok nesreče glede na določene atribute, ki smo jih dobili s sprobavanjem vseh kombinacij (uspelo nam je za max 5 atributov, saj je bilo časovno zahtevno)
Za meritve smo uporabljali funkcijo accuracy_score, ki nam pove točnost modela. Uspelo nam je izbrati atribute, ki so nam podali točnost 57%. 
Najboljši model za točnost so naključni gozdovi. Za ROC pa Gradient boost algoritem.
Najslabši je bil logistični. ROC grafov je kar nekaj, tukaj je najbolje ocenjen.



Za časovno napoved smo uporabili algoritem Sezonski ARIMA (SARIMA). <br>
Uporablja tri atribute (ki smo jih nastavili ročno, glede na točnost MAPE (Mean absolute percentage error) in obliku napovedi): 
* AR, ki predstavlja koliko preteklih vrednosti sezonskih obdobij, ki vplivajo na trenutno vrednost, mi smo uporabili AR = 5.
* I, ki je red integracije, katerega smo nastavili na 0 (nismo uporabili).
* MA, ki predstavlja vpliv preteklih napak sezonskega obdobja, mi smo ga nastavali na 3.

Nastavili smo tako, da smo za obdobje 2016 do 2022 testirali napovedi. Napovedi smo testirali z MAPE ter vizualno. Vizualno smo preverjali tako, da tudi, če ni so odstopanja, da je oblika podobna tisti, ki odstopa za vsako leto. 
Ko smo pridobili rezultate s katerimi smo bili zadovoljni, smo zaključili z izbiro atributov.
Pridobili smo MAPE = 8.765%, ki pove, koliko procentov odstopajo napovedi od dejanskih meritev. <br>
Napovedovali smo za leto 2023 in tudi 2024 ter to prikazali vizualno. Za leto 2023 je napovedanih 13298 nesreč (vsota).
